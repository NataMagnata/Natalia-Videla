```python
import PySAM.Pvwattsv8 as pv
import pandas as pd
import matplotlib.pyplot as plt

# === Ruta del archivo limpio ===
ruta_csv = "limpio_1554203_-19.95_130.50_tmy-2020.csv"  # Cámbialo por el que quieras usar

# === Parámetros generales ===
capacidades_kw = [500, 1000, 5000]

resultados = {}

for cap in capacidades_kw:
    sistema = pv.default("PVWattsNone")  # <-- CAMBIO AQUÍ
    sistema.SolarResource.solar_resource_file = ruta_csv
    sistema.SystemDesign.system_capacity = cap
    sistema.SystemDesign.dc_ac_ratio = 1.2
    sistema.Lifetime.analysis_period = 1
    sistema.execute()
    
    energia = sistema.Outputs.ac_annual
    resultados[cap] = energia

# === Graficar resultados ===
plt.figure(figsize=(8,5))
plt.bar([f"{c} kW" for c in capacidades_kw], [resultados[c] for c in capacidades_kw])
plt.ylabel("Energía anual generada (kWh)")
plt.title("Simulación de producción anual")
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
import PySAM.Pvwattsv8 as pv
import matplotlib.pyplot as plt

# === Archivos CSV por país ===
archivos = {
    "Australia": "limpio_1554203_-19.95_130.50_tmy-2020.csv",
    "Chile": "limpio_6064053_-28.44_-68.73_2022.csv",
    "EEUU": "limpio_594946_40.85_-100.78_tmy-2022.csv"
}

# === Parámetros comunes del sistema ===
system_capacity = 1000  # kW (puedes cambiar a 500 o 5000 para otras comparaciones)
dc_ac_ratio = 1.2
array_type = 1          # 0: Open Rack, 1: Roof Mount
tilt = 20               # grados
azimuth = 180           # sur

# === Resultados
resultados = {}

# === Simular para cada país ===
for pais, ruta_csv in archivos.items():
    sistema = pv.default("PVWattsCommercial")
    sistema.SolarResource.solar_resource_file = ruta_csv
    sistema.SystemDesign.system_capacity = system_capacity
    sistema.SystemDesign.dc_ac_ratio = dc_ac_ratio
    sistema.SystemDesign.array_type = array_type
    sistema.SystemDesign.tilt = tilt
    sistema.SystemDesign.azimuth = azimuth
    sistema.Lifetime.analysis_period = 1
    sistema.execute()

    energia = sistema.Outputs.ac_annual
    resultados[pais] = energia

# === Mostrar resultados
for pais, energia in resultados.items():
    print(f"{pais}: {energia:.2f} kWh/año")

# === Graficar comparativa
plt.figure(figsize=(8,5))
plt.bar(resultados.keys(), resultados.values(), color=["orange", "green", "blue"])
plt.ylabel("Energía anual generada (kWh)")
plt.title(f"Comparativa de producción anual - Capacidad {system_capacity} kW")
plt.grid(axis="y")
plt.tight_layout()
plt.show()
```

```python
import PySAM.Pvwattsv8 as pv
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# === Archivos CSV por país ===
archivos = {
    "Australia": "limpio_1554203_-19.95_130.50_tmy-2020.csv",
    "Chile": "limpio_6064053_-28.44_-68.73_2022.csv",
    "EEUU": "limpio_594946_40.85_-100.78_tmy-2022.csv"
}

# === Capacidades (kW) y colores personalizados
capacidades_kw = [500, 1000, 5000]
colores = {
    500: "#1f77b4",     # azul
    1000: "#ff7f0e",    # naranja
    5000: "#2ca02c"     # verde
}

# === Parámetros del sistema
dc_ac_ratio = 1.2
array_type = 1
tilt = 20
azimuth = 180

# === Simulaciones
resultados = []

for pais, ruta in archivos.items():
    for capacidad in capacidades_kw:
        sistema = pv.default("PVWattsCommercial")
        sistema.SolarResource.solar_resource_file = ruta
        sistema.SystemDesign.system_capacity = capacidad
        sistema.SystemDesign.dc_ac_ratio = dc_ac_ratio
        sistema.SystemDesign.array_type = array_type
        sistema.SystemDesign.tilt = tilt
        sistema.SystemDesign.azimuth = azimuth
        sistema.Lifetime.analysis_period = 1
        sistema.execute()

        energia = sistema.Outputs.ac_annual
        resultados.append({"País": pais, "Capacidad (kW)": capacidad, "Energía (kWh)": energia})

df = pd.DataFrame(resultados)
print(df)

# === Preparar gráfico de barras agrupadas
fig, ax = plt.subplots(figsize=(10, 6))

x = np.arange(len(archivos))  # posiciones de los países
ancho = 0.25  # ancho de barra

for i, cap in enumerate(capacidades_kw):
    energias = df[df["Capacidad (kW)"] == cap]["Energía (kWh)"].values
    ax.bar(x + i*ancho, energias, width=ancho, label=f"{cap} kW", color=colores[cap])

# Etiquetas y ajustes
ax.set_xticks(x + ancho)
ax.set_xticklabels(list(archivos.keys()))
ax.set_ylabel("Energía Anual Generada (kWh)")
ax.set_title("Comparativa de Producción Anual por País y Capacidad")
ax.legend(title="Capacidad de Planta")
ax.grid(axis="y")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import ticker
import PySAM.Pvwattsv8 as pv

# === Parámetros ===
ruta_csv = "limpio_1554203_-19.95_130.50_tmy-2020.csv"
capacidad_kw = 1000  # 1 MW
ratios = [1.0, 1.2, 1.4]
perdidas = [10, 14, 18]

# === Simulación ===
resultados = []

for ratio in ratios:
    for loss in perdidas:
        sistema = pv.default("PvWattsNone")
        sistema.SolarResource.solar_resource_file = ruta_csv
        sistema.SystemDesign.system_capacity = capacidad_kw
        sistema.SystemDesign.dc_ac_ratio = ratio
        sistema.Lifetime.analysis_period = 1
        sistema.SystemDesign.losses = loss  # % total de pérdidas
        sistema.execute()
        
        energia = sistema.Outputs.ac_annual
        resultados.append({
            "DC/AC Ratio": ratio,
            "Pérdidas (%)": loss,
            "Energía Anual (kWh)": energia
        })

df = pd.DataFrame(resultados)

# === Mostrar tabla sin notación científica ===
print(df.to_string(index=False, float_format="{:,.0f}".format))

# === Heatmap ===
pivot_table = df.pivot(index="Pérdidas (%)", columns="DC/AC Ratio", values="Energía Anual (kWh)")
plt.figure(figsize=(8, 6))
sns.heatmap(pivot_table, annot=True, fmt=".0f", cmap="YlGnBu", cbar_kws={'format': '%.0e'})

# Ajustar formato de la barra de color (cbar)
plt.gca().collections[0].colorbar.ax.yaxis.set_major_formatter(
    ticker.FuncFormatter(lambda x, _: f"{int(x):,}")
)

plt.title("Producción Anual (kWh) - 1 MW en Chile")
plt.xlabel("DC/AC Ratio")
plt.ylabel("Pérdidas (%)")
plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# === Leer el CSV ===
ruta_csv = "limpio_6064053_-28.44_-68.73_2022.csv"
df = pd.read_csv(ruta_csv, skiprows=2)

# === Crear columna de fecha ===
df['Fecha'] = pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour', 'Minute']])

# === Graficar en subplots ===
fig, axs = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

# --- GHI ---
axs[0].plot(df['Fecha'], df['GHI'], color='gold')
axs[0].set_ylabel("GHI (W/m²)")
axs[0].set_title("Irradiancia Global (GHI)")

# --- Temperatura ---
axs[1].plot(df['Fecha'], df['Temperature'], color='orange')
axs[1].set_ylabel("Temperatura (°C)")
axs[1].set_title("Temperatura del Aire")

# --- Velocidad del viento ---
axs[2].plot(df['Fecha'], df['Wind Speed'], color='blue')
axs[2].set_ylabel("Velocidad del Viento (m/s)")
axs[2].set_xlabel("Fecha")
axs[2].set_title("Velocidad del Viento")

# --- Ajustes generales ---
for ax in axs:
    ax.grid(True)

fig.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# === Definir rutas de archivos CSV ===
archivos = {
    'Australia': 'limpio_1554203_-19.95_130.50_tmy-2020.csv',
    'Chile': 'limpio_6064053_-28.44_-68.73_2022.csv',
    'EEUU': 'limpio_594946_40.85_-100.78_tmy-2022.csv'
}

# === Función para cargar y graficar ===
def procesar_y_graficar(pais, ruta_csv):
    df = pd.read_csv(ruta_csv, skiprows=2)

    # Crear columna de fecha
    df['Fecha'] = pd.to_datetime(df[['Year', 'Month', 'Day', 'Hour', 'Minute']])

    # === Gráficos ===
    fig, axs = plt.subplots(3, 1, figsize=(14, 10), sharex=True)

    # GHI
    axs[0].plot(df['Fecha'], df['GHI'], color='gold')
    axs[0].set_ylabel('GHI (W/m²)')
    axs[0].set_title(f'Irradiancia Global (GHI) - {pais}')
    axs[0].grid(True)

    # Temperatura
    axs[1].plot(df['Fecha'], df['Temperature'], color='orange')
    axs[1].set_ylabel('Temperatura (°C)')
    axs[1].set_title('Temperatura del Aire')
    axs[1].grid(True)

    # Velocidad del Viento
    axs[2].plot(df['Fecha'], df['Wind Speed'], color='blue')
    axs[2].set_ylabel('Velocidad del Viento (m/s)')
    axs[2].set_title('Velocidad del Viento')
    axs[2].set_xlabel('Fecha')
    axs[2].grid(True)

    plt.tight_layout()
    plt.show()

# === Procesar cada país ===
for pais, ruta in archivos.items():
    print(f"\nProcesando {pais}...")
    procesar_y_graficar(pais, ruta)
```
